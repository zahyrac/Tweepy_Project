```python
# Import packages
import os
import tweepy
import pandas as pd
```

```python
# Replace the keys you saved from the Twitter API at the appropriate locations

auth = tweepy.auth.OAuthHandler('INSERTKEY', 'INSERTKEY')
auth.set_access_token('INSERTKEY', 'INSERTKEY')

# avoid error
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
# parameters
search_words = "AppleEvent, AirPods3"
new_search = search_words + " -filter:retweets"
num_of_tweets = 100

# keyword search 
tweets = tweepy.Cursor(api.search_tweets,
                       q=new_search,
                       lang="en").items(num_of_tweets)

# list of tweets
tweet_text = [tweet.text.encode('utf-8') for tweet in tweets]

# view the list of tweets
tweet_text
```

```python
# predefined 4 parameters
tweets = tweepy.Cursor(api.search_tweets,
              q=new_search,
              lang="en").items(num_of_tweets)
# create tweets_info as lists of tweets information
tweets_info = [[tweet.id,
               tweet.user.screen_name, 
                tweet.user.followers_count, 
                tweet.user.location, 
                tweet.text.encode('utf-8')] for tweet in tweets]

# create a dataframe with column names
tweet_info = pd.DataFrame(data=tweets_info, 
                    columns=["Tweet ID","Screen Name", "Number of Followers", "Location", "Tweet"])
# view the tweet info dataframe
tweet_info
```

```python
# save the tweet info 
tweet_info.to_csv(search_words+"_tweets.csv",index=False)
```
